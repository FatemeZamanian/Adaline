```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
N=100
x_male=np.random.normal(10,2,N)
x_fmale=np.random.normal(50,4,N)
X=np.concatenate((x_fmale,x_male))
label_fmale=np.zeros(N,dtype='int')
label_male=np.ones(N,dtype='int')
Y=np.concatenate((label_fmale,label_male))
```

```python
plt.figure()
plt.scatter(X[0:N],Y[0:N], c='blue')
plt.scatter(X[N:],Y[N:], c='orange')

plt.xlabel('Hair')
plt.ylabel('Gender')
plt.show()
```

```python
X=X.reshape(200,1)
Y=Y.reshape(200,1)
for i in range(5):
    m=np.matmul(np.linalg.inv(np.matmul(X.T,X)),np.matmul(X.T,Y))+i
    y_pred=np.matmul(X,m)
    plt.scatter(X[0:N],Y[0:N], c='blue')
    plt.scatter(X[N:],Y[N:], c='orange')
    plt.plot(X,y_pred,c="red",lw=4)
```

```python
for i in range(5):
    m=np.random.random(1)
    y_pred=np.matmul(X,m)
    plt.scatter(X[0:N],Y[0:N], c='blue')
    plt.scatter(X[N:],Y[N:], c='orange')
    plt.plot(X,y_pred,c="red",lw=4)
```

```python
m = np.matmul(np.linalg.inv(np.matmul(X.T, X)),  np.matmul(X.T, Y))
y_pred=np.matmul(X,m)
plt.plot(X,y_pred,c="red",lw=4)

plt.scatter(X[0:N],Y[0:N], c='blue')
plt.scatter(X[N:],Y[N:], c='orange')
```
