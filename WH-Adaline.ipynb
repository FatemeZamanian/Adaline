```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
x, y, coef = datasets.make_regression(n_samples=400,n_features=1,n_informative=1,noise=20,coef=True,random_state=0)

x = np.interp(x, (x.min(), x.max()), (30,150))
y = np.interp(y, (y.min(), y.max()), (90, 210))

plt.ion()
plt.plot(x,y,'.',label='training data')
plt.xlabel('height');plt.ylabel('weigh')
plt.title('Height Vs. Weight')
```

```python
import random
N=400
std=0.5
temp=[]
x=np.random.rand(N)*100+100
for i in range(400):
    temp.append(random.randint(18,35))
y=np.divide(x,100)*np.divide(x,100)*temp
plt.ion()
plt.plot(x,y,'.',label='training data')
plt.xlabel('height');plt.ylabel('weigh')
plt.title('Height Vs. Weight')
x=x.reshape(400,1)
y=y.reshape(400,1)
```

```python
for i in range(5):
    m=np.matmul(np.linalg.inv(np.matmul(x.T,x)),np.matmul(x.T,y))+i
    y_pred=np.matmul(x,m)
    plt.scatter(x,y,c="blue")
    plt.plot(x,y_pred,c="red",lw=4)
```

```python
for i in range(5):
    m=np.random.rand(1)*i
    y_pred=np.matmul(x,m)
    plt.scatter(x,y,c="blue")
    plt.plot(x,y_pred,c="red",lw=4)
```

```python
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
plt.plot(x, intercept + slope*x, 'r', label='fitted line')
plt.legend()
plt.show()
```
